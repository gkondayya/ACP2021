```python
# importing Future data for different expiry and option data for different expiry and multiple strike price
from datetime import date
```

```python
s=date(2021,12,31)
nifty_fut = get_history(symbol='NIFTY', start=s,end=s,index=True,futures=True,expiry_date=s)
niftydata = nifty_fut
nifty_opt = get_history(symbol='NIFTY', start=s,end=s,index=True,option_type='CE',strike_price=16000,expiry_date=s)
optiondata=nifty_opt
s= s+dt.timedelta(days=1)
expiry = [date(2022,1,27),date(2022,2,24)]
for x in expiry:
  nifty_fut = get_history(symbol='NIFTY', start=s,end=x,index=True,futures=True,expiry_date=x)
  niftydata=niftydata.append(nifty_fut)
  high=nifty_fut['Close'].max()
  low=nifty_fut['Close'].min()
  # Rounding off high to nearest 100s
  high = int((round(high/100)*100)+100)
  low = int((round(low/100)*100)+100)

  for z in range(low,high,100):
    nifty_opt = get_history(symbol='NIFTY', start=s,end=x,index=True,option_type='CE',strike_price=z,expiry_date=x)
    optiondata = optiondata.append(nifty_opt)
    nifty_opt = get_history(symbol='NIFTY', start=s,end=x,index=True,option_type='PE',strike_price=z,expiry_date=x)
    optiondata = optiondata.append(nifty_opt)
  s = x+dt.timedelta(days=1)



optiondata.tail()


 
```
