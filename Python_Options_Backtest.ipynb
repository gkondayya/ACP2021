```python
import pandas as pd
```

```python
data = {"Stocks":["INFY","TCS","REL","AWL"],
        "Price":[1600,2500,3900,620]}
df = pd.DataFrame(data,columns=["Stocks","Price"])
df
```

```python
# Adding new column
df['Quantity'] =0
df
```

```python
df.columns
```

```python
# Selecting the last two values of Price
price_last2=df['Price'][-2:]
price_last2
```

```python
df.Price.min()
```

```python
df.Price.max()
```

```python
df.loc[0,'Quantity']=10      #this command will insert value ‘10’ in 1 st row of column ‘Quantity’
df.loc[1,'Quantity']=20
df.loc[2,'Quantity']=30
df.loc[3,'Quantity']=40
df
```

```python
df['Amount']=df['Price']*df['Quantity']
df
```

```python
df['CumSum']=df['Amount'].cumsum()
df
```

```python
df.drop(['CumSum'],axis=1,inplace=True)
df
```

```python
# Functions using Data Frame 
def df_func(df):
  try:
    return(df['Price']*df['Quantity'])
  except:
    return np.NaN

df['Amount']=df.apply(df_func,axis=1)
df

```
